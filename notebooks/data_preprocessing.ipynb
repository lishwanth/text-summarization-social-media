# Data Preprocessing and Exploration
This notebook is used for data exploration and preprocessing of the CNN/DailyMail dataset.

```
from datasets import load_dataset
import pandas as pd

# Load the dataset
dataset = load_dataset('cnn_dailymail', '3.0.0', split='train')

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(dataset)

# Save a sample to the data/raw folder
df[['article', 'highlights']].to_csv('../data/raw/articles.csv', index=False)

# Display the first few rows
df.head()
```
